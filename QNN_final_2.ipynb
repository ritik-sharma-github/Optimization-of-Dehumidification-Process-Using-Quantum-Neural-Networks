```python
import pandas as pd
import numpy as np
from scipy.stats.mstats import winsorize

# Assuming 'df' is your DataFrame containing the dataset
# Replace 'df' with your actual DataFrame name

# Define the columns
columns = ['L (mm)', 'd (mm)', 'x (mm)', 'Twi (℃)', 'Tai (℃)', 'Wai (g/kg)', 'ṁa (kg/s)', 'tc (s)', 'Two (℃)', 'Tao (℃)', 'Wao (g/kg)']

# Define winsorization limits
winsor_limits = 0.05

# Dictionary to store outlier counts for each feature
outlier_counts = {}

# Count outliers for each feature
for col in columns:
    if col in ['L (mm)', 'd (mm)', 'x (mm)', 'Tai (℃)', 'Wai (g/kg)', 'tc (s)', 'Two (℃)', 'Tao (℃)', 'Wao (g/kg)']:
        try:
            # Winsorize the column
            winsorized_values = winsorize(df[col], limits=(0, winsor_limits))
            # Count outliers
            outlier_count = ((df[col] > winsorized_values.max()) | (df[col] < winsorized_values.min())).sum()
            outlier_counts[col] = outlier_count
        except Exception as e:
            print(f"Error handling {col}: {str(e)}")

    elif col in ['Twi (℃)', 'ṁa (kg/s)']:
        try:
            # Check for NaN or negative values
            valid_values = df[col].dropna()
            valid_values = valid_values[valid_values > 0]  # Assuming all values should be positive

            # Log transform the column
            transformed_values = np.log(valid_values)
            # Count outliers
            outlier_count = (transformed_values > transformed_values.mean() + 3 * transformed_values.std()).sum()
            outlier_counts[col] = outlier_count
        except Exception as e:
            print(f"Error handling {col}: {str(e)}")

# Print outlier counts for each feature
for col, count in outlier_counts.items():
    print(f"Feature {col}: Outlier Count {count}")
```

```python
def remove_outliers(df, columns):
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    return df

# Remove outliers from the specified columns
df_no_outliers = remove_outliers(df, columns)
df_no_outliers
```

```python
def cap_outliers(df, columns):
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df[col] = np.where(df[col] < lower_bound, lower_bound, df[col])
        df[col] = np.where(df[col] > upper_bound, upper_bound, df[col])
    return df

# Cap outliers in the specified columns
df = cap_outliers(df, columns)
df
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.base import BaseEstimator, RegressorMixin
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
import pennylane as qml
import tensorflow as tf
import matplotlib.pyplot as plt

# Load your dataset and preprocess as needed
# Example assuming `df` contains your dataset with features and target variables

# Define features and target variables
X = df[['δd (mm)', 'L (mm)', 'd (mm)', 'x (mm)', 'Twi (℃)', 'Tai (℃)', 'Wai (g/kg)', 'ṁa (kg/s)', 'tc (s)']]
y = df[['Two (℃)', 'Tao (℃)', 'Wao (g/kg)']]

# Normalize the data
scaler_X = StandardScaler()
scaler_y = StandardScaler()

X_scaled = scaler_X.fit_transform(X)
y_scaled = scaler_y.fit_transform(y)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.2, random_state=42)

# Set up the quantum device
n_qubits = 9  # Number of qubits matching the number of features
dev = qml.device('default.qubit', wires=n_qubits)

# Define a quantum layer
@qml.qnode(dev, interface='tf')
def quantum_layer(inputs, weights):
    qml.templates.AngleEmbedding(inputs, wires=range(n_qubits))
    qml.templates.StronglyEntanglingLayers(weights, wires=range(n_qubits))
    return [qml.expval(qml.PauliZ(i)) for i in range(n_qubits)]

weight_shapes = {'weights': (3, n_qubits, 3)}

# Define a custom RMSE loss function
def rmse(y_true, y_pred):
    return tf.sqrt(tf.reduce_mean(tf.square(y_true - y_pred)))

# Build the individual QNN model
def build_qnn_model():
    model = Sequential()
    model.add(Dense(9, activation='relu', input_shape=(X_train.shape[1],)))
    model.add(Dense(9, activation='relu'))
    model.add(qml.qnn.KerasLayer(quantum_layer, weight_shapes, output_dim=9))
    model.add(Dense(3))  # Three outputs for Two, Tao, Wao
    return model

# Custom wrapper class for QNN model to make it compatible with BaggingRegressor
class KerasQNNRegressor(BaseEstimator, RegressorMixin):
    def __init__(self, model):
        self.model = model
        self.rmse_history = []

    def fit(self, X, y):
        self.model.compile(optimizer=Adam(learning_rate=0.001), loss=rmse)
        history = self.model.fit(X, y, epochs=100, batch_size=64, verbose=0)
        self.rmse_history.append(history.history['loss'])  # Store training RMSE history
        return self

    def predict(self, X):
        return self.model.predict(X)

# Build an ensemble of QNN models using BaggingRegressor
num_models = 5  # Number of QNN models in the ensemble
bagging_models = [KerasQNNRegressor(build_qnn_model()) for _ in range(num_models)]

# Fit each model individually and handle validation separately if needed
for model in bagging_models:
    model.fit(X_train, y_train)

# Evaluate the ensemble performance
y_pred = np.zeros_like(y_test)  # Accumulate predictions from each model
for model in bagging_models:
    y_pred += model.predict(X_test)
y_pred /= len(bagging_models)  # Average predictions from all models

# Calculate metrics for the ensemble
rmse_scores = np.sqrt(mean_squared_error(y_test, y_pred, multioutput='raw_values'))
mae_scores = mean_absolute_error(y_test, y_pred, multioutput='raw_values')
r2_scores = r2_score(y_test, y_pred, multioutput='raw_values')

print(f"RMSE Scores: {rmse_scores}")
print(f"MAE Scores: {mae_scores}")
print(f"R-squared Scores: {r2_scores}")

# Further analysis or tuning based on ensemble performance can be performed here

# Predict values using the trained ensemble model
new_data = X_test  # Replace with new data if available
predicted_values = np.zeros_like(new_data)
for model in bagging_models:
    predicted_values += model.predict(new_data)
predicted_values /= len(bagging_models)  # Average predictions from all models

# Inverse transform the predictions to original scale
predicted_values_original_scale = scaler_y.inverse_transform(predicted_values)

print(predicted_values_original_scale)

# Plot training loss values for each QNN model in the ensemble
for i, model in enumerate(bagging_models):
    plt.figure(figsize=(8, 6))
    plt.plot(model.rmse_history[0], label=f'Train RMSE Model {i+1}')
    plt.title(f'Model {i+1} Training Loss')
    plt.xlabel('Epoch')
    plt.ylabel('RMSE Loss')
    plt.legend()
    plt.grid(True)
    plt.show()
```
